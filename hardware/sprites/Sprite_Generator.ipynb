```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
directory = '/content/drive/MyDrive/SENIOR CAPSTONE/Sprites/Sprite_Images/'

for filename in os.listdir(directory):
    print(directory + '../Sprite_text' + filename.split('.')[0] + '.txt')

    # Open Image 
    image = Image.open(directory + filename)

    # Writing the output file for System Verilog
    with open(directory + '../Sprite_text/' + filename.split('.')[0] + '.txt', 'a') as f:
        for line in range(image.height):
            f.write(str(int(np.log(image.width)/np.log(2)))+"'d"+str(line)+" : pattern <= {")
            
            for col in range(image.width):
                pixel = image.getpixel((line,col))
                r = str(hex(pixel[0]*256*256 + pixel[1]*256 + pixel[2]))
                r = r[2:]
                f.write("{24'h"+r+"}, ")
                if col == image.width - 1:
                    f.write("{24'h"+r+"}};\n")

plt.imshow(image)
```
